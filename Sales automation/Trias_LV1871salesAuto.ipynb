```python
import pandas as pd
import numpy as np 
```

```python
df1= pd.read_excel('triassamplesales.xlsx')
```

```python
df18= df1.groupby(["Zertifikatsnummer","Periode auf Bestellebene"]).first()
df18.to_excel('groupingby.xlsx')
df18	
```

```python
#converting the datatype of dates in our data to data_time data type
df1['Periodenbeginn des Versicherungsschutzes']= pd.to_datetime(df1['Periodenbeginn des Versicherungsschutzes'])
df1['Periodenende des Versicherungsschutzes']= pd.to_datetime(df1['Periodenende des Versicherungsschutzes'])
df1['Periodenbeginn des Versicherungsschutzes']= pd.to_datetime(df1['Periodenbeginn des Versicherungsschutzes'])
df1['Periodenende des Versicherungsschutzes']= pd.to_datetime(df1['Periodenende des Versicherungsschutzes'])
# Calculating the difference b/w the following dates. If negatives difference is present then we have an error.
df1['Differenceperiod_biginn_end'] = df1['Periodenende des Versicherungsschutzes']-df1['Periodenbeginn des Versicherungsschutzes']
#Checking for days which are in negative so that we can find out the error in risikoend, risikostart and Vertragsbeginn, Vertragsende 
df1[df1['Differenceperiod_biginn_end']<'0 days']
```

```python
my_series = pd.Series(df1.Differenceperiod_biginn_end)
df1['Differenceperiod_biginn_end'] = my_series.to_frame()
print(df1['Differenceperiod_biginn_end'])
print(type(df1['Differenceperiod_biginn_end']))

risk= pd.DataFrame(df1['Differenceperiod_biginn_end'].dt.days)
def change_color(value):
  color= 'red' if value <0 else 'green'
  return 'color: %s' %color
df2= risk.style.applymap(change_color)
df2.to_excel('Differenceperiod_biginn_end.xlsx')
df2= pd.read_excel('Differenceperiod_biginn_end.xlsx')
df1["Differenceperiod_biginn_end"] = df2["Differenceperiod_biginn_end"]
```

```python
df1['Buchungsdatum']= pd.to_datetime(df1['Buchungsdatum'])
df3= df1[df1['Periodenbeginn des Versicherungsschutzes'] < df1['Buchungsdatum']]
df3.to_excel('datedif.xlsx')
df3
```

```python
df4= pd.read_excel('datedif.xlsx')
df1["Buchungdatum which are > period_beginn(Zertifikat NO.)"] = df4["Zertifikatsnummer"]
df1.to_excel('kaufdatum.xlsx')
df1
```
