```python
import pandas as pd
import numpy as np 
```

```python
df1= pd.read_excel('AXA_sales_09.22.xlsx')
```

```python
#highlighting the duplicates in policy number number
from pandas.core.indexes.multi import duplicated
HK= df1[df1['Policy Number'].duplicated()]
dupli= HK.style.highlight_null('red')
dupli.to_excel('dupli.xlsx')
dupli
```

```python
df3= pd.read_excel('dupli.xlsx')
df1["Duplicates_inPolicynumber"] = df3["Policy Number"]
df1.head()
```

```python
df1['Policy Start Date']= pd.to_datetime(df1['Policy Start Date'])
df1['Policy End Date']= pd.to_datetime(df1['Policy End Date'])
df1['DifferencePolicy_START_END'] = df1['Policy End Date']-df1['Policy Start Date']
#Checking for days which are in negative so that we can find out the error in policy start and end 
df1[df1['DifferencePolicy_START_END']<'0 days']
```

```python
my_series = pd.Series(df1.DifferencePolicy_START_END)
df1['DifferencePolicy_START_END'] = my_series.to_frame()
print(df1['DifferencePolicy_START_END'])
print(type(df1['DifferencePolicy_START_END']))
```

```python
risk= pd.DataFrame(df1['DifferencePolicy_START_END'].dt.days)
def change_color(value):
  color= 'red' if value <0 else 'green'
  return 'color: %s' %color
```

```python
df2= risk.style.applymap(change_color)
df2.to_excel('DifferencePolicy_START_END.xlsx')
```

```python
df2= pd.read_excel('DifferencePolicy_START_END.xlsx')
df1["DifferencePolicy_START_END(Check policy no.)"] = df2["DifferencePolicy_START_END"]
df1.head(61)
```
