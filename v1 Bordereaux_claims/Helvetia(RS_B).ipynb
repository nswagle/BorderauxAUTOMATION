```python
#converting the datatype of dates in our data to data_time data type
df1['Risikostart']= pd.to_datetime(df1['Risikostart'])
df1['Risikoende']= pd.to_datetime(df1['Risikoende'])
df1['Vertragsbeginn gem. Zertifikat']= pd.to_datetime(df1['Vertragsbeginn gem. Zertifikat'])
df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']= pd.to_datetime(df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende'])
df1['Risikostart']= pd.to_datetime(df1['Risikostart'])
df1['Risikoende']= pd.to_datetime(df1['Risikoende'])
df1['Vertragsbeginn gem. Zertifikat']= pd.to_datetime(df1['Vertragsbeginn gem. Zertifikat'])
df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']= pd.to_datetime(df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende'])

# Calculating the difference b/w the following dates. If negatives difference is present then we have an error.
df1['DifferenceRisk'] = df1['Risikoende']-df1['Risikostart']
df1['Differencevertrag'] = df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']-df1['Vertragsbeginn gem. Zertifikat']
#Checking for days which are in negative so that we can find out the error in risikoend, risikostart and Vertragsbeginn, Vertragsende 
df1[df1['Differencevertrag']<'0 days']
```

```python
my_series = pd.Series(df1.DifferenceRisk)
df1['DifferenceRisk'] = my_series.to_frame()
print(df1['DifferenceRisk'])
print(type(df1['DifferenceRisk']))
my_series = pd.Series(df1.Differencevertrag)
df1['Differencevertrag'] = my_series.to_frame()
print(df1['Differencevertrag'])
print(type(df1['Differencevertrag']))
```

```python
risk= pd.DataFrame(df1['DifferenceRisk'].dt.days)
vertrag= pd.DataFrame(df1['Differencevertrag'].dt.days)
def change_color(value):
  color= 'red' if value <0 else 'green'
  return 'color: %s' %color
df2= risk.style.applymap(change_color)
df3= vertrag.style.applymap(change_color)
df2.to_excel('Rdifference.xlsx')
df3.to_excel('Vdifference.xlsx')
```
