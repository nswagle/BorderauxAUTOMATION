```python
import numpy as np
import pandas as pd
```

```python
# imported libraries and now importing the file
df1= pd.read_csv('newlvclaim.csv')
```

```python
#converting the datatype of dates in our data to data_time data type
df1['Risikostart']= pd.to_datetime(df1['Risikostart'])
df1['Risikoende']= pd.to_datetime(df1['Risikoende'])
df1['Vertragsbeginn gem. Zertifikat']= pd.to_datetime(df1['Vertragsbeginn gem. Zertifikat'])
df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']= pd.to_datetime(df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende'])
```

```python
df1['Risikostart']= pd.to_datetime(df1['Risikostart'])
df1['Risikoende']= pd.to_datetime(df1['Risikoende'])
df1['Vertragsbeginn gem. Zertifikat']= pd.to_datetime(df1['Vertragsbeginn gem. Zertifikat'])
df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']= pd.to_datetime(df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende'])

# Calculating the difference b/w the following dates. If negatives difference is present then we have an error
df1['DifferenceRisk'] = df1['Risikoende']-df1['Risikostart']
df1['Differencevertrag'] = df1['Vertragsende gem. Zertifikat/voraussichtliches Vertragsende']-df1['Vertragsbeginn gem. Zertifikat']
```

```python
df4= df1[df1['DifferenceRisk']<'0 days']
df4.to_excel('DifferenceRisk.xlsx')
```
